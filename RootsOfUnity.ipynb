```python
import numpy as np
import matplotlib.pyplot as plt
from cmath import phase
```

```python
x1d = np.linspace(-2, 2, 400)
y1d = np.linspace(-2, 2, 400)
x, y = np.meshgrid(x1d, y1d)
z = x + 1j*y

def NR(myArray):
    for i in range(20):
        myArray = 2/3*myArray + 1/(3*(myArray**2))
    return myArray

final =np.vectorize(phase)(NR(z))

fractal = plt.imshow(final, extent=(-2,2, -2,2), cmap = 'plasma')
plt.colorbar(fractal)
plt.xlabel("Re")
plt.ylabel("Im")
    
```
