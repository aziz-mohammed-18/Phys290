```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def Pi(N):
    x = np.random.uniform(0, 1, N)
    return np.mean(np.sqrt(1-x**2))

def m(x, y):
    return(np.mean(x*y)-np.mean(x)*np.mean(y))/(np.mean(x**2)-(np.mean(x)**2))
```

```python
N = np.linspace(100, 10000, 100)

Nf = []
sigmaf = []
for n in N:
    resultlist = []
    for i in range(100):
        resultlist.append(Pi(int(n)))
    Nf.append(n)
    sigmaf.append(np.std(resultlist))
plt.plot(Nf, sigmaf)
plt.show()
plt.plot(np.log(Nf), np.log(sigmaf))
plt.show()
m(np.log(Nf), np.log(sigmaf))
```
