```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
a_list = []
x_list = []
num_its = 100


def LogMap(a, x):
    return a*x*(1-x)

for a in np.linspace(0., 4., 1000):
    # if a%1 == 0:
    # print("a =", a)
    for x in np.linspace(0., 1., 200):
        for i in range(num_its): x = LogMap(a, x)
        a_list = np.append(a_list, x)
        x_list = np.append(x_list, a)
# for i in range(len(a_list)):
#     print("a =", a_list[i], " x = ", x_list[i])

plt.scatter( x_list, a_list, marker = '.', s = 0.01)
plt.xlabel("a")
plt.ylabel("x")
plt.title("Logistic Map: \n Furry Bicycles on Vacation (Bifurcations)")
plt.show()
```
