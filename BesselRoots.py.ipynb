```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.special import * 
```

```python
def roughroots(num_roots):
    x = 0.0
    rroots = []
    while(len(rroots)<num_roots):
        if (j0(x)*j0(x-0.1)) < 0:
            rroots.append(x)
        x += 0.1
    return np.array(rroots)

def polish(ar, maxError):
    # er = ar
    # while er.any() > maxError:
        # er = abs(j0(ar)/j1(ar))
    for i in range(10):
        ar = ar + j0(ar)/j1(ar)

    return ar
```

```python
rr = roughroots(20)
pr = polish(rr, 1.e-3)

x = np.linspace(0, 70, 700)
y = np.zeros(700)

for i in range(700):
    y[i] = (j0(i/10))

plt.plot(x,y, color = 'blue', label="$J_0(x)$", linewidth = 1)
plt.scatter(rr, j0(rr), color= 'hotpink', marker = 'o', label = "Rough Roots")
plt.scatter(pr, np.zeros(20), color = 'purple', marker = 'x', label = "Polished Roots")
plt.xlabel("x")
plt.legend()
plt.show()
```
